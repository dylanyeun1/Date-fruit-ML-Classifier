```python
#libraries
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

df = pd.read_excel('C:\\Users\dylan\Python Projects\ML\Date_Fruit_Datasets\Date_Fruit_Datasets.xlsx')
print(df)
```

```python
#Replacing date names w/ numerical values
name_dict = {'BERHI':1, 'DEGLET':2, 'DOKOL': 3, 'IRAQI': 4, 'ROTANA': 6, 'SAFAVI': 7, 'SOGAY': 8}
df['Name'] = df['Class'].map(name_dict)
df.head()
```

```python
x = df[['AREA', 'PERIMETER', 'MAJOR_AXIS', 'MINOR_AXIS', 'ECCENTRICITY', 'EQDIASQ', 'SOLIDITY', 'CONVEX_AREA', 'EXTENT', 'ASPECT_RATIO']]
y = df['Name']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
#prediction based train data
predictions = clf.predict([[422163, 2378.9080, 837.8484, 645.6693, 0.6373, 733.1539, 0.9947, 424428, 0.7831, 1.2976]])
print(predictions)
#probable outcomes
possibilities = clf.predict_proba([[422163, 2378.9080, 837.8484, 645.6693, 0.6373, 733.1539, 0.9947, 424428, 0.7831, 1.2976]])
print(possibilities)
```

```python
#Test
testx = clf.predict(x_test)
print(testx)
print(y_test)
```

```python
#Score
#80-82% accuracy
clf.score(x_test, y_test)
```
